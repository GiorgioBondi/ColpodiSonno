```python
%matplotlib inline
from pynq.pmods import Grove_IMU
from IPython import display
from ipywidgets import widgets
import numpy as np
import scipy.signal as signal
import time
import matplotlib.pyplot as plt


# initialize Grove IMU on PMOD 1 (JB) and grove id 4 (GR4).
# PMOD mapping is: PMOD 1 -> JB, PMOD 2 -> JC, PMOD 3 -> JD, PMOD 4 -> JE
# NOTE: only GR4 can be used for the Grove IMU
imu = Grove_IMU(1,4)

#empty lists creation
accX=list()
accY=list()
accZ=list()
pitch_l=list()

while (True):

        # get data from accelerometer
        accl = imu.get_accl()

        accX.append(accl[0])
        accY.append(accl[1])
        accZ.append(accl[2])
        
        #calculate pitch from accelerations
        argomento= accl[1]/np.sqrt(accl[0]*accl[0]*accl[1]*accl[1])
        pitch = np.arctan(argomento)
        pitch_l.append(pitch)
        
        time.sleep(0.1)
        
    except KeyboardInterrupt:
        # exit without showing errors if the stop button in Jupyter is pressed
        break

#array creation from data-filled lists
accX_arr=np.array(accX)
accY_arr=np.array(accY)
accZ_arr=np.array(accZ)
pitch_arr=np.array(pitch_l)

np.savetxt("DataXmov.npy", accX_arr)
np.savetxt("DataYmov.npy", accY_arr)
np.savetxt("DataZmov.npy", accZ_arr)
np.savetxt("DataPmov.npy",pitch_arr)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataZ=np.loadtxt("DataZ.npy")
movAvgZ=pd.Series(dataZ).rolling(len(dataZ)*0.1).mean()

#calcolo della Standard Deviation per identificazione outlier
movStdZ=pd.Series(dataZ).rolling(len(dataZ)*0.1).std()

for i in range(len(dataZ)):
    if ((dataZ[i]-movAvgZ[i])>2*movStdZ[i]):
        print("Al secondo "+str(i*0.1)+" c'è stato un improvviso movimento del capo!")
        plt.axvline(x=0.1*i)
#plot
t=np.arange(0,len(dataZ)*0.1,0.1)
plt.figure(figsize=(10,10))
plt.plot(t,dataZ,color='red')
plt.plot(t,movAvgZ,color='blue')
plt.show()
```

```python
#cross-correlazione tra finestra con zero-padding e segnale rilevat
#funziona ma riportando ogni elemento di finestre come lista in entrata 
#al template
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.signal as signal
from numpy.fft import rfft, irfft

dataZ=np.loadtxt("DataZ.npy")

print(len(dataZ))

#le due liste da confrontare devono avere stessa lunghezza

template=list()
template=[ 0.08,0.13,0.11,0.19,0.31,1.5,0.66,0.68 ,0.88,1.03]
    
#Zero padding
for i in range (10,300):
    template.append(0)

# Fast Fourier Transform (FFT)
dataZ = rfft(dataZ)

# Inverto il vettore template
tempY = []
while len(template) > 0:
    tempY.append(template.pop())
y = rfft(tempY)

# Faccio la formula inversa della FFT
CrossCorrelazione = irfft(dataZ*y)
print(len(CrossCorrelazione))

   
x=np.arange(0,300)
plt.plot(x,CrossCorrelazione)
plt.xlabel("Campioni")
plt.ylabel("CrossCorrelazione")
line1=plt.plot(x,[3]*300,'r')
plt.show()


```
